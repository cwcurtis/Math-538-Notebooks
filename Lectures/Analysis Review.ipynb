# Maps and Analysis

So while this course is not necessarily focused on rigor, it is unavoidable that at various points in the course we will need to make recourse to several core concepts from analysis.  That being said, most of what we will need can be framed entirely within one dimension,and so unless very clearly, and very explicitly stated otherwise, we can do almost everything we will need with respect to one variable.  

With that being said, let me explain the entire course in about one line.  Suppose we have a function $f(x)$, where $f:\mathbb{R}\rightarrow \mathbb{R}$.  Pick an initial point $x_{0}$.  Define a sequence $\left\{x_{n}\right\}_{n=0}^{\infty}$ via the equation

$$
x_{n} = f(x_{n-1}), ~ n\geq 1.
$$

> **Central Question of Course**:  What is the behavior of $x_{n}$ for large values of $n$?

Note, we do not even need to make many requirements on $f$ in this class.  To wit, we define the circle map $C_{\alpha,n}(x)$ via the equation

$$
C_{\alpha,n}(x) = \alpha x \mbox{mod}(n), ~ n \in \mathbb{Z}^{+}.
$$

What we mean by this is that $C_{\alpha,n}$ maps the real line $\mathbb{R}$ to the interval $[-n,n]$, where at every point $x_{j} = jn/\alpha$, the function "resets" to zero.  Thus, for example the circle map $C_{2,1}(x)$ looks like 

![circ_map](https://upload.wikimedia.org/wikipedia/commons/6/68/Exampleergodicmap.svg)

Note, this is called a cirlce because while the map appears discontinuous when plotted on the real line, we can think of it as a continuous helix wrapping around a circle of perimeter $n/\alpha$.  Thus, you will also see this map desribed as 

$$
C_{\alpha,n} : \mathbb{R}/( (n/\alpha)\mathbb{Z}) \rightarrow [-n,n].
$$

where the periodicity of the function is encapsulated in the expression $\mathbb{R}/( (n/\alpha)\mathbb{Z})$, which simply means that we repeat the function over interals $n/\alpha$ long.  

Of particular interest in this course are the family of _ unimodal _ maps.  

> **Unimodal Map**: A function $f$ is said to be unimodal if there exists a closed bounded interval $[a,b]$ such that $f:[a,b]\rightarrow[a,b]$, $f(a)=f(b)$, and there exists a unique point $c\in(a,b)$ called the _ turning point _ such that $f$ is _ strictly increasing _ on $[a,c)$ and is _ strictly decreasing _ on $(c,b]$. Note, by $f:[a,b]\rightarrow [a,b]$, we mean only that the image of $f$ over $[a,b]$, is contained within $[a,b]$.  It need not always equal $[a,b]$.

Note, the _ strictly increasing _ and _ strictly decreasing _ is not strictly necessary (hehe, couldn't help it), but we will never look at an example in this course where this is not the case, so I am making life a little easier here.  

_ Problem _: Show via linear transformations and scalings that you lose no generality by supposing that a unimodal map is defined via the requirements

$$
f:[0,1]\rightarrow[0,1], ~ f(0)=f(1)=0,
$$

and $c\in(0,1)$.

Examples of this that we will talk about at length in this course are _ Tent Maps _, $T_{\mu}(x)$ where 

$$
T_{\mu}(x) = \left \{ 
\begin{array}{rl}
\mu x & x \leq \frac{1}{2} \\
\mu(1 - x) & \frac{1}{2} < x .
\end{array}
\right.
$$

![Tent_Map](https://upload.wikimedia.org/wikipedia/commons/a/a3/Tent_map_2.png)

Thus we see that $T_{\mu}(x)$ will be unimodal for $0\leq\mu\leq2$, where $T_{\mu}:[0,1]\rightarrow[0,1]$ with turning point $c=1/2$.  

An even more important example of a unimodal map is that of the logistic map $L_{\alpha}(x)$ where 

$$
L_{\alpha}(x) = \alpha x(1-x).
$$

_ Problem _: Determine what values of $\alpha$ and over what intervals the logistic map is unimodal.   

So what is it about unimodal maps that makes them so interesting?  Well, remember the central question of the course.  Suppose we have a unimodal map $f:[0,1]\rightarrow[0,1]$, and we choose $x_{0}\in [0,1]$. Then we necessarily see that the entire sequence generated from this choice must always stay in $[0,1]$, i.e. 

$$
x_{n} \in [0,1], ~ x_{n} = f(x_{n-1}) = f^{(n)}(x_{0}), ~ n\geq 1.
$$

Thus, the sequence $\left\{x_{n}\right\}_{n=0}^{\infty}$ is trapped in the closed, bounded interval $[0,1]$.  This motivates us to remind ourselves of the definition of a compact set.  

>**Compact Set**: A set $E\subset \mathbb{R}$ is compact if for any sequence $x_{n}\in E$, there exists a subsequence $x_{n_{j}}$ which converges to a point in $E$.

Of course, we would like some means of then characterizing in advance when a subset of the real line may or may not be compact.  To do this, we first need to remind ourselves of the formal definitions of open and closed sets on the real line.  

>**$\epsilon$-interval**: For $x\in\mathbb{R}$, we define the $\epsilon$-interval around $x$ to be $I_{\epsilon}(x)=\left\{y\in\mathbb{R}:|x-y|<\epsilon\right\}$. 

>**Open Set**: A set $E\subset \mathbb{R}$ is open if for every $x\in E$, there is some $\epsilon>0$ such that $I_{\epsilon}(x)\in E$. 

>**Closed Set**: A set $E\subset \mathbb{R}$ is closed if its complement, $E^{c}=\mathbb{R}\backslash E=\left\{y\in\mathbb{R}:y\notin E\right\}$ is open.  

Let's briefly remind ourselves of some key facts about open and closed sets.  The first is 

> Some sets are neither open nor closed.  See (a,b], [a,b) for example.  


> Arbitrary unions of open sets are open, i.e. if $O_{j}$, $j=0,1,\cdots$ are open, then so is $\cup_{j}O_{j}$.


> Arbitrary intersections of closed sets are closed, i.e. if $C_{j}$, $j=0,1,\cdots$ are closed, then so is $\cap_{j} C_{j}$.


> Finite intersections of open sets are open.  Finite unions of closed sets are closed.  But when we take an infinite limit in either case, we cannot say whether the limiting set is open or closed.  

On the real line, we can couple this definition with the following theorem.

>**Bolzano-Weierstrass Theorem** - A set $E\subset \mathbb{R}$ is compact if and only if it is closed and bounded.
