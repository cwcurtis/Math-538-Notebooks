# Maps and Analysis

So while this course is not necessarily focused on rigor, it is unavoidable that at various points in the course we will need to make recourse to several core concepts from analysis.  That being said, most of what we will need can be framed entirely within one dimension,and so unless very clearly, and very explicitly stated otherwise, we can do almost everything we will need with respect to one variable.  

With that being said, let me explain the entire course in about one line.  Suppose we have a function $f(x)$, where $f:\mathbb{R}\rightarrow \mathbb{R}$.  Pick an initial point $x_{0}$.  Define a sequence $\left\{x_{n}\right\}_{n=0}^{\infty}$ via the equation

$$
x_{n} = f(x_{n-1}), ~ n\geq 1.
$$

> **Central Question of Course**:  What is the behavior of $x_{n}$ for large values of $n$?

Note, we do not even need to make many requirements on $f$ in this class.  To wit, we define the circle map $C_{\alpha,n}(x)$ via the equation

$$
C_{\alpha,n}(x) = \alpha x \mbox{mod}(n), ~ n \in \mathbb{Z}^{+}.
$$

What we mean by this is that $C_{\alpha,n}$ maps the real line $\mathbb{R}$ to the interval $[-n,n]$, where at every point $x_{j} = jn/\alpha$, the function "resets" to zero.  Thus, for example the circle map $C_{2,1}(x)$ looks like 

![circ_map](https://upload.wikimedia.org/wikipedia/commons/6/68/Exampleergodicmap.svg)

Note, this is called a cirlce because while the map appears discontinuous when plotted on the real line, we can think of it as a continuous helix wrapping around a circle of perimeter $n/\alpha$.  Thus, you will also see this map desribed as 

$$
C_{\alpha,n} : \mathbb{R}/( (n/\alpha)\mathbb{Z}) \rightarrow [-n,n].
$$

where the periodicity of the function is encapsulated in the expression $\mathbb{R}/( (n/\alpha)\mathbb{Z})$, which simply means that we repeat the function over interals $n/\alpha$ long.  

Of particular interest in this course are the family of _ unimodal _ maps.  

> **Unimodal Map**: A function $f$ is said to be unimodal if there exists a closed bounded interval $[a,b]$ such that $f:[a,b]\rightarrow[a,b]$, $f(a)=f(b)$, and there exists a unique point $c\in(a,b)$ called the _ turning point _ such that $f$ is _ strictly increasing _ on $[a,c)$ and is _ strictly decreasing _ on $(c,b]$. Note, by $f:[a,b]\rightarrow [a,b]$, we mean only that the image of $f$ over $[a,b]$, is contained within $[a,b]$.  It need not always equal $[a,b]$.

Note, the _ strictly increasing _ and _ strictly decreasing _ is not strictly necessary (hehe, couldn't help it), but we will never look at an example in this course where this is not the case, so I am making life a little easier here.  

_ Problem _: Show via linear transformations and scalings that you lose no generality by supposing that a unimodal map is defined via the requirements

$$
f:[0,1]\rightarrow[0,1], ~ f(0)=f(1)=0,
$$

and $c\in(0,1)$.

Examples of this that we will talk about at length in this course are _ Tent Maps _, $T_{\mu}(x)$ where 

$$
T_{\mu}(x) = \left \{ 
\begin{array}{rl}
\mu x & x \leq \frac{1}{2} \\
\mu(1 - x) & \frac{1}{2} < x .
\end{array}
\right.
$$

![Tent_Map](https://upload.wikimedia.org/wikipedia/commons/a/a3/Tent_map_2.png)

Thus we see that $T_{\mu}(x)$ will be unimodal for $0\leq\mu\leq2$, where $T_{\mu}:[0,1]\rightarrow[0,1]$ with turning point $c=1/2$.  

An even more important example of a unimodal map is that of the logistic map $L_{\alpha}(x)$ where 

$$
L_{\alpha}(x) = \alpha x(1-x).
$$

_ Problem _: Determine what values of $\alpha$ and over what intervals the logistic map is unimodal.   

So what is it about unimodal maps that makes them so interesting?  Well, remember the central question of the course.  Suppose we have a unimodal map $f:[0,1]\rightarrow[0,1]$, and we choose $x_{0}\in [0,1]$. Then we necessarily see that the entire sequence generated from this choice must always stay in $[0,1]$, i.e. 

$$
x_{n} \in [0,1], ~ x_{n} = f(x_{n-1}) = f^{(n)}(x_{0}), ~ n\geq 1.
$$

Thus, the sequence $\left\{x_{n}\right\}_{n=0}^{\infty}$ is trapped in the closed, bounded interval $[0,1]$.  This motivates us to remind ourselves of the definition of a compact set.  

>**Compact Set**: A set $E\subset \mathbb{R}$ is compact if for any sequence $x_{n}\in E$, there exists a subsequence $x_{n_{j}}$ which converges to a point in $E$.

Of course, we would like some means of then characterizing in advance when a subset of the real line may or may not be compact.  To do this, we first need to remind ourselves of the formal definitions of open and closed sets on the real line.  

>**$\epsilon$-interval**: For $x\in\mathbb{R}$, we define the $\epsilon$-interval around $x$ to be $I_{\epsilon}(x)=\left\{y\in\mathbb{R}:|x-y|<\epsilon\right\}$. 

>**Open Set**: A set $E\subset \mathbb{R}$ is open if for every $x\in E$, there is some $\epsilon>0$ such that $I_{\epsilon}(x)\in E$. Note, in effect, we have that any open set in $\mathbb{R}$ is a union of different $I_{\epsilon}(x)$ subsets.  

>**Closed Set**: A set $E\subset \mathbb{R}$ is closed if its complement, $E^{c}=\mathbb{R}\backslash E=\left\{y\in\mathbb{R}:y\notin E\right\}$ is open.  

>**Accumulation Point**: A point $x$ is an accumulation point of a subset $E$ if for all $\epsilon > 0$, $E \cap (I_{\epsilon}\backslash\left\{x\right\})\neq \emptyset$.

>**Closure of a Subset**: Let $E'$ be the set of accumulation points of $E$.  We call $\bar{E}=E\cup E'$ the closure of $E$.

>**Boundary of a Set**: We define the boundary of a subset $E\subset \mathbb{R}$ to be $\partial E = \bar{E}\cap \bar{E^{c}}$.   

>**Connected Subsets**: Let $E\subset \mathbb{R}$.  Define an open subset _ relative _ to $E$ to be any set $E\cap O$, where $O$ is an open set in $\mathbb{R}$ (i.e. a collection of open intervals).  We say that $E$ is connected if it cannot be written as the disjoint union of two relatively open subsets, i.e. we cannot write $E=H\cup K$, where $H$ and $K$ are relatively open in $E$ and $H \cap K = \emptyset$.

>**Continuous Functions**: A function $f:\mathbb{R}\rightarrow\mathbb{R}$ is continuous at a point $x$ if for all $\epsilon>0$ there exists a $\delta>0$ such that if $|x-y|<\delta$, then $|f(x)-f(y)|<\epsilon$. 

Let's briefly remind ourselves of some key facts associated with these statements.  The first is 

> Some sets are neither open nor closed.  See (a,b], [a,b) for example.  


> Arbitrary unions of open sets are open, i.e. if $O_{j}$, $j=0,1,\cdots$ are open, then so is $\cup_{j}O_{j}$.


> Arbitrary intersections of closed sets are closed, i.e. if $C_{j}$, $j=0,1,\cdots$ are closed, then so is $\cap_{j} C_{j}$.


> Finite intersections of open sets are open.  Finite unions of closed sets are closed.  But when we take an infinite limit in either case, we cannot say whether the limiting set is open or closed.  

> If $E$ is closed then $E=\bar{E}$, and the set of accumulation points of $E$ is identical to the boundary of $E$.  If $E$ is open, then again its set of accumulation points is identical to its boundary.  


> It is frustrating that the definition of a connected set is so long-winded.  But it does make sense.  For example, you should be able to see why $E=[1,3/2)\cup(3/2,2]$ is disconnected.  If I use as my open sets in $\mathbb{R}$ $(1-\epsilon,3/2)$ and $(3/2,2+\epsilon)$, where $\epsilon>0$, then we see that relative to $E$,the two disjoint sets which we used to define $E$ are open, and therefore we can write $E$ as the union of disjoint open sets.  

>Likewise, if we let $E=[1,2]$, then if we supposed that $E$ were disconnected, that would mean there were two disjoint relatively open sets $H$ and $K$ such that $[1,2]=H\cup K$, which is to say every point in $[1,2]$ was either in $H$ or $K$.  The sets being relatively open though means that there are sequences within both $H$ and $K$ which must converge to points not in the set, but $[1,2]$ is closed, and therefore we have a contradiction.     

>So, we have a couple of equivalent statements of continuity.  The most important one is that if $f$ is continuous, then if $x_{n}\rightarrow x$, then $f(x_{n})\rightarrow f(x)$.  The other is a bit weirder, and says that if $f(x)\in E$, and $E$ is open, then there is an open subset $U$ such that $x\in U$ and $f(U)\subset E$.  We will use the first statement constantly.  The second one, not as often, but see the problems below.    

_ Problem _: Show that the set of accumulation points $E'$ of a subset $E$ is closed.  

_ Problem _: Show that $\bar{E}$ is closed.

_ Problem _: Show that for a closed set $E$, $E\backslash \partial E$ is open.  This set is called the interior of $E$.  

_ Problem _: Prove all of the unproven above statements.  See also if you can construct a closed set from the intersection of an infinite number of open sets.  

_ Problem _: Let $E$ be a closed subset, let $x_{n}\in E$, and suppose that $x_{n}\rightarrow x$.  Show that $x\in E$.

_ Problem _: Show that if a subset $E$ is compact and $f$ is continuous, then $f(E)$ is also compact.  

_ Problem _: Show that if a subset $E$ is connected and $f$ is continuous, then $f(E)$ is connected.  

On the real line, we now can better understand the following theorem, which is a central result we will come back to again and again throughout the semester.  

>**Bolzano-Weierstrass Theorem** - A set $E\subset \mathbb{R}$ is compact if and only if it is closed and bounded.

Thus we already see a rather interesting result with regards to unimodal maps.  In effect, unimodal maps ensure that any iterative sequence necessarily has a convergent subsequence in trapping interval $[0,1]$.  What is worth keeping in mind is that we do not know without far more work what the convergence subsequence is, or if there is only one.  As we will show, unimodal maps are able to generate _ very _ complicated sequences which do not have ready characterizations.  

On this note, it is worth thinking about more general notions of how sequences $x_{n}$ behave in the large $n$-limit.  To wit, we define

\begin{align*}
\limsup_{n \rightarrow \infty} x_{n} = & \inf_{n\geq 0}\sup_{m\geq n}x_{m},\\ 
\liminf_{n \rightarrow \infty} x_{n} = & \sup_{n\geq 0}\inf_{m\geq n}x_{m}
\end{align*}

Note, a _ supremum _ or $\sup$ is a _ least-upper-bound _.  This, takes into account both obvious upper bounds as well as limits.  Thus, if $x_{n}$ is the sequence

$$
x_{n} = \left\{\begin{array}{rl} -1 & n~\mbox{even}\\ 1 & n~\mbox{odd}\end{array}\right.
$$

then $\sup_{n\geq 0}x_{n} = 1$.  If $x_{n}$ is the sequence 

$$
x_{n} = \frac{n+1}{n+2}
$$

by noting that 

$$
x_{n} = \frac{n+2-1}{n+2} = 1 - \frac{1}{n+2},
$$

then $\sup_{n\geq 0}x_{n} = 1$.  

So, while it seems obvious that we have for $n\geq 0$ that 

$$
x_{n} \leq \sup_{m\geq 0} x_{m},
$$

the _ least-upper-bound _ part means that for every $\epsilon > 0$, there exists a $N$ such that

$$
\sup_{m \geq 0} x_{m} < x_{N} + \epsilon.
$$

Thus the supremum is that upper bound that is _ just _ bigger than every element of the sequence. 

Likewise, an _ infimum _ or $\inf$ is a _ greatest-lower-bound _.  Thus we have for $n \geq 0$ that 

$$
x_{n} \geq \inf_{m\geq 0} x_{m},
$$

but, for every $\epsilon>0$, there exists a $\tilde{N}$ such that 

$$
x_{\tilde{N}} < \inf_{m\geq 0}x_{m} + \epsilon.
$$

Thus, the infimum of a sequence is that value that is _ just _ smaller than every member of the sequence. Again, this allows us to include limits so that if we have 

$$
x_{n} = \frac{n+2}{n+1} = \frac{n+1+1}{n+1} = 1 + \frac{1}{n+1}, 
$$

then $\inf_{n\geq0}x_{n} = 1$.

Roughly speaking, the $\limsup$ is the smallest upper bound for the sequence in the large $n$ limit, while the $\liminf$ is the largest lower bound in the large $n$ limit.  Note, either can diverge to $\pm \infty$.  

_ Problem _: For $n\geq0$, let $x_{n}$ be the sequence

$$
x_{n} = \left\{\begin{array}{rl} -1 & n~\mbox{even}\\ 1 & n~\mbox{odd}\end{array}\right.
$$

Show that 

$$
\limsup_{n\rightarrow\infty}x_{n} = 1, ~ \limsup_{n\rightarrow\infty}x_{n}=-1.
$$

To see this smallest upper and biggest lower bound in the large quality, solve:

_ Problem _: For $n\geq0$, let $\tilde{x}_{n}$ be the sequence such that $\tilde{x}_{n}=10$ for $n\leq 23$, and $\tilde{x}_{n}$ is given by 

$$
\tilde{x}_{n} = \left\{\begin{array}{rl} -1 & n~\mbox{even}\\ 1 & n~\mbox{odd}\end{array}\right.
$$

for $n>23$.  Show that 

$$
\limsup_{n\rightarrow\infty}\tilde{x}_{n} = 1, ~ \limsup_{n\rightarrow\infty}\tilde{x}_{n}=-1.
$$

So we see that the $\limsup$ and $\liminf$ are able to ignore transitory behavior in a sequence and instead capture the large $n$ phenomena.  Now, as is readily apparent, we always have 

$$
\liminf_{n\rightarrow\infty}x_{n} \leq \limsup_{n\rightarrow\infty}x_{n}.
$$

Where it gets interesting is if they are equal, in which case we have that if $\limsup_{n\rightarrow\infty}x_{n} = \liminf_{n\rightarrow\infty}x_{n}$, and these values are finite, then 

$$
\lim_{n\rightarrow\infty} x_{n} = \limsup_{n\rightarrow\infty}x_{n} = \liminf_{n\rightarrow\infty}x_{n}.
$$

Thus, we see that a sequence has a well defined limit when its smallest upper bound for large values of $n$ is the same as its largest lower bound for large values of $n$. To see this result, let $x^{\ast} = \limsup_{n\rightarrow\infty}x_{n} = \liminf_{n\rightarrow\infty}x_{n}$.  Thus, using the results about infimums and supremums for above, for all $\epsilon>0$, we must have an $N$ and $\tilde{N}$ such that

$$
\sup_{m\geq\tilde{N}}x_{m} < x^{\ast} + \epsilon, ~ x^{\ast} < \inf_{m\geq N}x_{m} + \epsilon.
$$

Let $\bar{N} = \max\left\{N,\tilde{N}\right\}$, which then gives us 

$$
x_{m} < x^{\ast} + \epsilon, ~ x^{\ast} < x_{m} + \epsilon, ~ m\geq \bar{N},
$$

which is the same thing as saying

$$
\left|x_{m} - x^{\ast}\right| < \epsilon, ~ m\geq \bar{N}.
$$

Since $\epsilon>0$ was arbitrary, we have thus shown 

$$
\lim_{m\rightarrow\infty}x_{m} = x^{\ast} = \limsup_{n\rightarrow\infty}x_{n} = \liminf_{n\rightarrow\infty}x_{n}.
$$
