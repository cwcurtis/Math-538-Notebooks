** Problem 2 ** - 5 pts: Show that the middle thirds Cantor set is perfect.  This means that the set is closed and is equal to the set of its accumulation points.  Again, a point $x$ is an accumulation point of a set $E$ if for all $\epsilon > 0$, we have 

$$
E \cap \left(I_{\epsilon}(x)\backslash\{x\}\right) \neq \emptyset,
$$

or $x$ is an accumulation point of $E$ if every neighborhood of $x$ contains a point of $E$ different from $x$ itself.  The trick to this is to use the fact that 

$$
\mathcal{C} = \bigcap_{j=0}^{\infty} C_{j}, 
$$

i.e every point in $\mathcal{C}$ has to come from some set of subintervals which remains after deletion.  Note, the hard part of the proof is starting from a point $x$ which you assume to be an accumulation point of $\mathcal{C}$, and then show that it necessarily must be in $\mathcal{C}$.
