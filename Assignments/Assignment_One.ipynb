$$

$$

$$

\left|\sigma_{n}\right|\leq & \frac{1}{n}\sum_{j=0}^{N}\left|a_{j}\right| + \frac{1}{n}\sum_{j=N+1}^{n}\left|a_{j}\right|,\\
\leq & \frac{1}{n}\sum_{j=0}^{N}\left|a_{j}\right| + \frac{n-N}{n}\frac{\epsilon}{2},\\
